### Цель исследования.

 **Исследование рынка российкого кинопроката и выявление текущих трендов.**
 
 ### Описание данных 
 
 **Таблица mkrf_movies** 
 

 `title` — название фильма;
 
`puNumber` — номер прокатного удостоверения;

`show_start_date` — дата премьеры фильма;

`type` — тип фильма;

`film_studio` — студия-производитель;

`production_country` — страна-производитель;

`director` — режиссёр;

`producer` — продюсер;

`age_restriction` — возрастная категория;

`refundable_support` — объём возвратных средств государственной поддержки;

`nonrefundable_support` — объём невозвратных средств государственной поддержки;

`financing_source` — источник государственного финансирования;

`budget` — общий бюджет фильма;

`ratings` — рейтинг фильма на КиноПоиске;

`genres` — жанр фильма.


 **Таблица mkrf_shows** 
 
`puNumber` — номер прокатного удостоверения;

`box_office` — сборы в рублях.

### Дополнительные столбцы

 - столбец с информацией о годе проката `year`
 - столбец с именем и фамилией главного режиссёра `director_name`

## Исследовательский анализ

### Количество фильмов в год

```python
title_pivot = df.pivot_table(index='year', values=('title','box_office'), aggfunc='count')
title_pivot.plot(y='title', kind='bar',title='Общее количество фильмов'); 
share_cinemas = df['box_office'].count()/df['title'].count()
print('Доля фильмов, с данными о прокате в кинотеатре:{:.0%}'.format(share_cinemas))
title_pivot['share_cinemas'] = title_pivot['box_office']/title_pivot['title']*100 
title_pivot.plot(y='share_cinemas', kind='bar',title='Доля фильмов с данными о прокате в кинотеатрах %');
title_pivot.plot(y=['title','box_office'],kind='bar',title='в кинотеатрах');
```

### Динамика проката по годам. Средняя и медианная сумма сборов для каждого года.

```python
box_office_pivot = df.pivot_table(index='year', values='box_office',aggfunc=['sum','mean','median'])
box_office_pivot.columns = ['sum','mean','median']
box_office_pivot.plot(y='sum', kind='bar',title='Суммы проката');
box_office_pivot.plot(y='mean', kind='bar',title='Сумма проката средняя');
box_office_pivot.plot(y='median', kind='bar',title='Сумма проката медианная');
box_office_pivot.style.highlight_max(color='yellowgreen', subset=['sum','mean','median'])\
.highlight_min(color='coral', subset=['sum','mean','median'])
```

*Минимальная сумма проката была в 2010 году, максимальная в 2017. Учитывая тот факт, что наши данные неполные, нашли средие значения проката в зависимости от количества данных в каждом году. 
Минимальные и максимальные значения так и осталиьс в 2010 и в 2017 годах, так же видно что в 2010-2013 года суммы сблоров были незначительны по сравнению с последующими годами. Но общий характер изменения итоговой суммы по годам и среднего значения немного отличается. Вокруг пика в 2017 общая сумма держадась примерно такой жже высокой в диапазоне 2016-2019 годов, тогда как по среднему значению видно, что средняя сумма сборов в 2016,2018 и 2019 годах была сильно ниже(так как количество фильмов вышедших в прокат в кинотеатрах в эти годы больше).
В медианных значениях, помимо общих черт со средними в 2010-2013 годы, есть различия: в 2014 году медианная сумма проката гораздо меньше средней, это говорито том, что был выброс с большой стоимостью( фильм с нетипично большими кассовыми сборами), и такая же ситуация видимо была в 2016 году.*
